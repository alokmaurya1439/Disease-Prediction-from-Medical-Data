# Understanding data


```python
class Analysis:
    def __init__(self, df):
        self.df = df

    def missing_value(self):
        return self.df.isnull().sum()
    
    def duplicate_value(self):
        return self.df.duplicated().sum()
    
    
data = Analysis(df)
```

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class EDA:
    def __init__(self, df):
        self.df = df

    def histogram(self, column):
        plt.hist(self.df[column], bins=10)
        plt.title(f'Histogram of {column}')
        plt.show

    def distplot(self,column):
        sns.kdeplot(self.df[column])
        plt.title(f'Distribution of {column}')
        plt.show

    def scatterplot(self, x_col, y_col):
        sns.scatterplot(x=x_col, y=y_col, data=self.df)
        plt.title(f'Scatterplot of {x_col} and {y_col}')
        plt.show
    
    def pairplot(self):
        sns.pairplot(self.df)
        plt.title('Pairplot of the data')
        plt.show

    def heatmap(self):
        plt.figure(figsize=(10,8))
        sns.heatmap(self.df.corr(), annot=True)
        plt.title("Heatmap of data")
        plt.show
        
eda = EDA(df)
```

```python
df = df.drop(columns=['trestbps', 'chol', 'fbs', 'restecg'])
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=['target'])
y = df['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled = StandardScaler()
x_train_scaled = scaled.fit_transform(x_train)
x_test_scaled = scaled.transform(x_test) 
```

```python
scaled
```

```python
pd.DataFrame(x_train_scaled)
```

```python
pd.DataFrame(x_test_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
lr = LogisticRegression()
lr.fit(x_train_scaled, y_train)
```

```python
svc = SVC()
svc.fit(x_train_scaled, y_train)
```

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
xg = XGBClassifier()
xg.fit(x_train, y_train)
```

```python
s_pred = svc.predict(x_test_scaled)
ac = accuracy_score(y_test, s_pred)
cm = confusion_matrix(y_test, s_pred)
cr = classification_report(y_test, s_pred)
```

```python
ac  , cm, cr
```

```python
r_pred = rf.predict(x_test)
ac = accuracy_score(y_test, r_pred)
cm = confusion_matrix(y_test, r_pred)
cr= classification_report(y_test, r_pred)
```

```python
ac  , cm, cr
```

```python
xg_pred = xg.predict(x_test)
ac = accuracy_score(y_test, xg_pred)
cm = confusion_matrix(y_test, xg_pred)
cr= classification_report(y_test, xg_pred)
```

```python
ac  , cm, cr
```

```python
y_pred = lr.predict(x_test)
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cr= classification_report(y_test, y_pred)
```
